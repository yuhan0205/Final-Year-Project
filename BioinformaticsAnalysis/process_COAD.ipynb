```python
import pandas as pd
```

```python
#select 1000 representative(top scores) of ~6000 cancer specific peaks
```

```python
coad_original = pd.read_csv(r'COAD_peakCalls_test.csv')
coad_original
```

```python
coad_eliminated = pd.read_csv(r'COAD_backHg38.bed', sep='\t', header=None)
coad_eliminated.columns = ["seqnames", "start", "end"]
coad_eliminated
```

```python
coad_merged = pd.merge(coad_original, coad_eliminated, on=["seqnames", "start"], how="inner")
coad_merged
```

```python
coad_merged = coad_merged[coad_merged.seqnames != 'chrY']
coad_merged
```

```python
coad_merged.annotation.value_counts()
```

```python
coad_merged.score.describe()
```

```python
coad_merged_top = coad_merged.sort_values(by='score',ascending=False).head(1000)
coad_merged_top
```

```python
coad_merged_bottom = coad_merged.sort_values(by='score',ascending=True).head(1000)
coad_merged_bottom
```

```python
coad_merged_top.annotation.value_counts()
```

```python
coad_merged_bottom.annotation.value_counts()
```

```python
coad_merged_top[["seqnames","start","end_x"]].to_csv('coad_top1000.bed', sep='\t', index=False, header=False)
```

```python
coad_merged_bottom[["seqnames","start","end_x"]].to_csv('coad_bottom1000.bed', sep='\t', index=False, header=False)
```

```python
#grouped by annotation
```

```python
adfs = {}
for annotation in list(coad_merged.annotation.value_counts().index):
    adf = coad_merged[coad_merged.annotation == annotation]
    adfs[annotation] = adf
    adf_output = adf[["seqnames","start","end_x"]]
    adf_output.to_csv("coad_"+annotation+".bed", sep='\t', index=False, header=False)
```

```python
adfs_top100 = {}
for annotation in list(coad_merged.annotation.value_counts().index):
    adf_top100 = coad_merged[coad_merged.annotation == annotation].sort_values(by="score", ascending=False).head(100)
    adfs_top100[annotation] = adf_top100
    top_output = adf_top100[["seqnames","start","end_x"]]
    top_output.to_csv("coad_"+annotation+"_top100.bed", sep='\t', index=False, header=False)
```

```python
adfs_top1000 = {}
for annotation in list(coad_merged_top.annotation.value_counts().index):
    adf_top1000 = coad_merged_top[coad_merged_top.annotation == annotation]
    adfs_top1000[annotation] = adf_top1000
    top_output = adf_top1000[["seqnames","start","end_x"]]
    top_output.to_csv("coad_"+annotation+"_intop1000.bed", sep='\t', index=False, header=False)
```
