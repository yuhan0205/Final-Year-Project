```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('C:\\Users\\mayh\\OneDrive - A STAR\\atac_seq\\model\\score\\apr25-COAD-h') as f:
    peaks_h = [line.rstrip('\n').replace(":","_") for line in f]
```

```python
len(peaks_h)
```

```python
num_list_h = ['556', '557', '558', '559', '560', '561', '562', '563', '564', '565', '566']
mean_dic = {}
std_dic = {}
avg_dic = {}
remove_list = []
for peak in peaks_h:
    dic = {}
    for num in num_list_h:
        fn = 'C:\\Users\\mayh\\OneDrive - A STAR\\atac_seq\\model\\score\\healthy\\nml-depth-WHC' + num + '-' + peak + '.txt'
        df = pd.read_csv(fn, header = None, sep = ' ')
        dic[num] = df[2]
    peak_df = pd.DataFrame(dic)
    mean = list(peak_df.mean(axis = 1))
    std = list(peak_df.std(axis = 1))
    if len(mean) == 2000 and len(std) == 2000:
        mean_dic[peak] = mean
        std_dic[peak] = std
    else:
        remove_list.append(peak)
```

```python
remove_list
```

```python
mean = pd.DataFrame(mean_dic)
std = pd.DataFrame(std_dic)
mean.to_csv('mean.csv')
std.to_csv('std.csv')
```

```python
for value in remove_list:
    value = value[3:]
    peaks_c.remove(value)
len(peaks_c)
```

```python
num_list_c = ['1014_180816', '1531_160616', '1531_180119', '519_210114', '809_030915', '809_110914', '986_100215', '986_261016']
dfs = {}
defects = {}
for num in num_list_c:
    dic = {}
    defect = []
    for peak in peaks_c:
        new_peak = 'chr' + peak
        fn = 'C:\\Users\\mayh\\OneDrive - A STAR\\atac_seq\\model\\score\\cancer\\nml-depth-' + num + '-' + peak + '.txt'
        df = pd.read_csv(fn, header = None, sep = ' ')
        if len(df[2]) != 2000:
            defect.append(peak)
            continue
        dic[new_peak] = df[2]
    defects[num] = defect
    num_df = pd.DataFrame(dic)
    dfs[num] = num_df
```
