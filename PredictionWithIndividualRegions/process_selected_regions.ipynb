```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('merged_score_selected.csv').iloc[:,:3]
df
```

```python
df['start'] = df['start']-5000
df['end'] = df['end']+5000
df
```

```python
df['region_h_altered'] = df['chr'].map(str)+'_'+df['start'].map(str)+'-'+df['end'].map(str)
df['region_c_altered'] = df['region_h_altered'].str[3:]
df
```

```python
df['region_h'] = df['chr']+':'+df['start'].map(str)+'-'+df['end'].map(str)
df['region_c'] = df['region_h'].str[3:]
df
```

```python
df['region_h_altered'].to_csv('merged_healthy_list_altered.txt', index=False, header=False)
df['region_c_altered'].to_csv('merged_cancer_list_altered.txt', index=False, header=False)
```

```python
df['region_h'].to_csv('merged_healthy_list.txt', index=False, header=False)
df['region_c'].to_csv('merged_cancer_list.txt', index=False, header=False)
```
